```python
class Producto():

#Define los atributos de la clase producto
  def __init__(self, identificacion, nombre, descripcion, cantidad, precio):
    self.identificacion = identificacion
    self.nombre = nombre
    self.descripcion = descripcion
    self.cantidad = cantidad
    self.precio = precio

#Obtiene la informacion de la clase y la imprime
  def obtener_info(self):
    print(f"Estos es la información del producto solicitado\n"
          " Nombre = ", self.nombre,
          "\n Descripción = ", self.descripcion,
          "\n Cantidad = ", self.cantidad,
          "\n Precio = ", self.precio)

#Cambia la info de un producto
  def cambiar_info(self):
    print("Qué información desea cambiar del producto, ingrese el número correspondiente a la acción que desee realizar")
    try:
      x = int(input(" 1.Cambiar nombre\n 2.Cambiar descripción\n 3.Cambiar cantidad\n 4.Cambiar precio\n"))
      if x == 1:
         self.nombre = input("Ingrese el nuevo nombre del producto\n")
      if x == 2:
       self.descripcion = input("Ingrese una nueva descripción del producto\n")
      if x == 3:
        try: #Como la cantidad tiene que ser un número este try lo verifica
         self.cantidad = int(input("Ingrese la nueva cantidad del producto\n"))
        except ValueError:
          print("La cantidad tiene que ser un número, por favor inténtelo nuevamente")
      if x == 4:
        try: #como el precio tiene qu ser un número este try lo verifica
         self.precio = float(input("Ingrese el nuevo precio del producto\n"))
        except ValueError:
          print("El precio tiene que ser un número, por favor inténtelo nuevamente")
      print("Información cambiada exitósamente")
    except ValueError:
     return ("Opción inválida inténtelo nuevamente")

#elementos de prueba para self.inventario, en este caso los productos tienen nombre diferente, pero
#esto no influye (o por lo menos no debería influir) en la ejecución del programa, solo están adentro para
#facilitar la comprobación de errores
bicicleta = Producto(123, "Bicicleta", "bicicleta ultimo modelo", 4, 500000)
polystation3 = Producto(222, "Polystation3", "Consola de última generacion", 4, 300000)

class Inventario():

#introducidos elementos de prueba a self.inventario
#si se desea iniciar el programa sin estos elementos simplemente elimínelos
    def __init__(self):
      self.inventario = {123:bicicleta,
                         222: polystation3
                         }


    def agregar_item(self):
      while True:
        try:
         identificacion = int(input("Ingrese la ID del producto\n")) #En este caso la ID del producto siempre sera un numero entero
         if identificacion not in self.inventario: #confirma que la id no está en el inventario para así poder agregarla
          break
         else: #Le informa al usuario que la id ya está en el sistema y le da la posibilidad de reemplazarla o cancelar el ingreso
           respuesta = input("La ID ingresada ya se encuentra en el sistema, si desea reemplazar"
          " el producto ingrese 'si', si desea intentar con otra ID ingrese cualquier caracter, si desea"
                             " cancelar la operacion ingrese 'cancelar'\n")
         if respuesta == "si":#reemplaza la id en el inventario
            break
         elif respuesta == "cancelar":#cancela la funcion
            return
        except ValueError:
          print("La ID tiene que ser un número entero, por favor inténtelo nuevamente")
      nombre = input("Ingrese el nombre del producto\n")
      descripcion = input("Ingrese la descripción del producto\n")
      while True:#Verifica que la cantidad ingresada sea un número entero
       try:
        cantidad = int(input("Ingrese la cantidad del producto\n"))
        break
       except ValueError:
        print("Cantidad ingresada debe ser un número entero por favor inténtelo nuevamente")
      while True:#Verifica que el precio ingresado sea un número
       try:
         precio = float(input("Ingrese el precio del producto\n"))
         break
       except ValueError:
         print("Precio tiene que ser un número por favor inténtelo nuevamente")
      producto = Producto(identificacion, nombre, descripcion, cantidad, precio)
      # Los productos son agregados al diccionario "inventario" con la llave correspondiente a su ID (en int), y el valor como
      # "producto", "producto" posee toda la información del objeto clase Producto, el problema es que todos los
      # productos se ingresaran con el mismo valor "producto" (pero con valor de llave "ID" diferente), el programa de todas
      #formas nunca llama a el valor de la llave solo identifica la id y desplega la información del producto
      # usando su función de la clase
      self.inventario.update({identificacion: producto})
      print("Producto agregado exitosamente")


    def eliminar_item(self):
      try:#comprueba que la id ingresada sea un número entero
       id = int(input("Ingrese la ID del producto que desea eliminar"))
       if id in self.inventario: #comprueba que la id se encuentre en el inventario
        eliminado = self.inventario.pop(id)
        print(f"La ID {id} con nombre {eliminado.nombre} ha sido eliminado del inventario exitosamente")
       else:
        print("ID ingresada no se encuentra en el inventario")
      except ValueError:
        print("ID debe ser un número entero")


    def actualizar_inventario(self):
      try:
       id = int(input("Ingrese la ID del producto que desea actualizar"))
       if id in self.inventario:
        return self.inventario[id].cambiar_info()
       else:
        return "ID ingresada no se encuentra en el inventario"
      except ValueError:
        print("ID tiene que ser un número entero, por favor inténtelo nuevamente")


    def buscar(self): # Error intended: ingresar algo distinto a número
     try:
      id = int(input("Ingrese la id del producto a buscar\n"))
      if id in self.inventario:
        self.inventario[id].obtener_info()
      else:
        print("ID ingresada no se encuentra en el inventario")
     except ValueError:
      print("Id tiene que ser un número, por favor inténtelo denuevo")



    def chequeo_inventario(self):
      cantidad_total = 0
      valor_total = 0
      print(f"Este es el estado del inventario :")
      for i in self.inventario:  #imprime toda la información de las "ID" en el inventario
        print(f" Nombre = {self.inventario[i].nombre}\n id = {self.inventario[i].identificacion}\n Descripción = {self.inventario[i].descripcion} \n"
              f" Cantidad = {self.inventario[i].cantidad}\n Precio = {self.inventario[i].precio} pesos")
        cantidad_total += int(self.inventario[i].cantidad)
        valor_total += (int(self.inventario[i].cantidad) * int(self.inventario[i].precio))
      print(f"El valor total del inventario es: {valor_total}")
      print(f"La cantidad total de productos en el inventario es {cantidad_total}")


print("Bienvenido al gestor de inventario")
print("Ingrese el número la opción que desee realizar")
bodega1 = Inventario()
while True: #mientras no se seleccione la acción 6 siempre se ejecutará esta "interfaz"
  try:
   opcion = int(input(f" 1. Ingresar un producto\n 2. Buscar un producto con su ID\n 3. Cambiar la información de un producto\n"
    f" 4. Eliminar un producto\n 5. Mostrar todos los productos del inventario\n 6. Si desea salir ingrese salir\n"))
   if opcion == 1:
      bodega1.agregar_item()
   elif opcion == 2:
      bodega1.buscar()
   elif opcion == 3:
      bodega1.actualizar_inventario()
   elif opcion == 4:
      bodega1.eliminar_item()
   elif opcion == 5:
     bodega1.chequeo_inventario()
   elif opcion == 6:
     break
   else:
    print("Opción inválida, por favor inténtelo nuevamente")
  except ValueError:
   print("Opción inválida, para ejecutar la acción debe ingresar algún número de la lista"
          " por favor ingresar nuevamente")






```
